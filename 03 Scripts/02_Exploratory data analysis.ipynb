### 3. Correlation heatmap

```python
# Creating a subset excluding columns that are not pertinent

sub = poetry[['Day of the week','Weather', 'Food', 'Sport', 'Meditation', 'Dance', 'Music', 'Guitar', 'Reading', 'Mood in the evening', 'sleep quality night before', 'mindfulness', 'totalwords all poems', 'no of poems per day']]
```

```python
# Creating a subplot with matplotlib

f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.

corr = sns.heatmap(sub.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
                                                      # place the correlation coefficients onto the heatmap.
```

### 7. Categorical plot

```python
# We are using a histogram to visualize the distribution of the variables. 
# This way, we can determine sensible categories for different ranges in the data. 
# The argument "kde" add a line that encompasses the distribution

sns.histplot(poetry['Meditation'], bins = 20, kde = True)
```

```python
# Creating a new column to categorize the scoring of "Meditation"

poetry.loc[poetry['Meditation'] < 30, 'meditation length category'] = 'short meditation'
```

```python
poetry.loc[(poetry['Meditation'] >= 30) & (poetry['Meditation'] <45), 'meditation length category'] = 'medium meditation'
```

```python
poetry.loc[poetry['Meditation'] > 40, 'meditation length category'] = 'long meditation'
```

```python
# Counting the values of the new column "meditation length category"

poetry['meditation length category'].value_counts(dropna = False)
```

```python
# Checking if the new "meditation length category" column is there
poetry.columns
```

```python
# Creating a categorical plot in seaborn using the meditation length categories created above and "Dance" as x

sns.set(style="ticks")
g = sns.catplot(x="Dance", y="totalwords all poems", hue="meditation length category", data=poetry)
```

So this gives me more insight into my data. It looks like I am a lazy dancer (because I only ever danced 10 minutes) but a strong meditator, because on days where I didn't dance I had medium and long meditations. At the same time, when I danced (even if it was just for 10 minutes...) I once wrote 30 words. I also once wrote between 30 and 40 words when I had only danced 5 minutes. In this plot you can also see that I mostly have long and medium medtations and that the highest wordcounts (between 50 and above 60 words) falls together whith long meditations.

### 8. Synopsis of the exploratory analysis

This exploratory analysis gave me more insight into my data. It makes sense to continue to test the hypothesis **"The more I meditate, the more words I write for my poems"**. Especially the cat(egorical) plot above made it clear that there is some kind of relationship between me meditating for a long or for a medium duration and the number of words I write for my poems. I will have to check though if it really is a linear relationship or if the relationship between the two variables "meditation" and "total number of words in my poems" is more complex.
